```python
def sort_func(row):
    hold = row['Work Description']
    if re.search('built up',hold,flags=re.I):
        return 'Built Up'
    elif re.search(r"class\s*\W*A\W*\s*\W*or\W*\s*\W*B\W*",hold,flags=re.I):
        return 'Class A or B'
    if re.search('partial',hold,flags=re.I):
        return 'Partial'
    if re.search('comp shingle',hold,flags=re.I):
        return 'Comp Shingle'
    else:
        return hold
    
```

```python
data['Reroof Type'] = data.apply(lambda row: sort_func(row), axis=1)
```

```python
def make_address(row):
    if str(row['Address Start']) != 'nan':
        hold = str(row['Address Start']) +' '
    if str(row['Street Direction']) != 'nan':
        hold += str(row['Street Direction']) +' '
    if str(row['Street Name']) != 'nan':
        hold += str(row['Street Name']) +' '
    if str(row['Street Suffix']) != 'nan':
        hold += str(row['Street Suffix']) +' '
    hold += 'Los Angeles CA, ' 
    if str(row['Zip Code']) != 'nan':
        hold += str(row['Zip Code'])
    return hold
```

```python
data.apply(lambda row: make_address(row), axis=1)
```

```python
data['Address'] = data.apply(lambda row: make_address(row), axis=1)
```

```python
clean_data = data.reset_index()[['Issue Date','License Expiration Date','Reroof Type','Address','Latitude/Longitude']]
```
